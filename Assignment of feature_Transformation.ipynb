# Log Trasformation

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
def log_transform(x):
    return np.log(x)
```

```python
log_transformer = FunctionTransformer(log_transform)
log_transformer
```

```python
log_transformer.fit(value)
```

```python
log_transformed_price = log_transformer.transform(value)
log_transformed_price
```

```python
plt.plot(value)
```

```python
plt.plot(log_transformed_price)
```
